```python
## SEGMENTING THE DATA TO 60s EPOCHS

def segment_fn(data, fs=128, size=60):
    sgmt_data = []
    sgmt = fs*size
    while(len(data) > sgmt):
        x = data[:sgmt]
        sgmt_data.append(x)
        data = data[sgmt:len(data)] 
    return sgmt_data

alz_sg = []
for i in range(15):
    x = segment_fn(alz_df[i])
    alz_sg.append(x)

mig_sg = []
for i in range(13):
    x = segment_fn(mig_df[i])
    mig_sg.append(x)

sei_sg = []
for i in range(10):
    x = segment_fn(sei_df[i])
    sei_sg.append(x)

nor_sg = []
for i in range(3):
    x = segment_fn(nor_df[i])
    nor_sg.append(x)
```

```python
# STATISTICAL FEATURES FUNCTIONS

# MEAN

def mean_fn(data):
    x = np.mean(data, axis = 0)
    return x

# VARIANCE

def var_fn(data):
    x = np.var(data, axis = 0)
    return x

# STANDARD DEVIATION

def std_fn(data):
    x = np.std(data, axis = 0)
    return x

# SKEWNESS

def skew_fn(data):
    x = skew(data, axis = 0)
    return x

# KURTOSIS 

def kurt_fn(data):
    x = kurtosis(data, axis = 0)
    return x

# PEAK MAGNITUDE
## FIGURE IT OUT LATER
```
